```python
data_csv['h-l_%']=(data_csv['adj_high']-data_csv['adj_low'])/data_csv['adj_low']*100
```

```python
data_csv['%_chng']=(data_csv['adj_close']-data_csv['adj_open'])/data_csv['adj_open']*100
#percentage change can be neagaive convert o positive
```

```python
df_features=data_csv[['adj_open','adj_close','adj_high','adj_low','adj_volume']]
df=data_csv[['adj_close','h-l_%','%_chng','adj_volume']]  # only this is used as features not data_csv
#data_csv=data_csv.drop(['date','ticker'],1)
#data_csv.drop(inplace=True)
data_csv.head(6)
```

```python
import math 
```

```python
#we wan to predict todays adj_close from previous 10% entries in adj_close
df.fillna(-9999,inplace=True)
forecast_col='adj_close'

forecast_out=int(math.ceil(0.001*(len(df))))
print (forecast_out)

df['label']=df[forecast_col].shift(-forecast_out)
```

```python
#df_temp=data_csv[['adj_open','adj_close','adj_high','adj_low','adj_volume','label']]
#df_temp[['adj_close','label']].head(101)
#we can see by shift we shifted adjclose up by 100 and stored it in label column
#we are comparing forcast price to adj_close price---- main 
#data_csv
```

```python
import numpy as np
```

```python
#x--features   y---labels

x=np.array(df.drop(['label'],1))
x_lately=x[-forecast_out:]
x=x[:-forecast_out]

df.dropna(inplace=True)
y=np.array(df['label'])

x=preprocessing.scale(x)

#actually we are comparing adj_close to label
```

```python

from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=cross_validation.train_test_split(x,y,test_size=0.3)
```

```python
clf=LinearRegression()
clf.fit(x_train,y_train)
accuracy=clf.score(x_test,y_test)
predict_accuracy=clf.predict(x_lately)  # how does classifier knows to return value in list or any other data structure?
```

```python
clf2=svm.SVR()
clf2.fit(x_train,y_train)
accuracy2=clf2.score(x_test,y_test)
```
