```python
import numpy as np
from sklearn import preprocessing,cross_validation,neighbors
import pandas as pd
```

```python
data_txt=pd.read_csv('/home/ishaan/Codes/Datasets/adult.data.txt')
print(type(data_txt))
```

```python
data_txt.replace('?',-99999,inplace=True)
#label=class_income
#features=everything except education number

print (type(data_txt['edu_num']))
data_txt.drop(['edu_num'],1,inplace=True)
#data_txt.drop(['country'],1,inplace=True)
```

```python
data_txt
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data_txt['workclass']=LabelEncoder().fit_transform(data_txt['workclass'])
data_txt['edu']=LabelEncoder().fit_transform(data_txt['edu'])
data_txt['marital_status']=LabelEncoder().fit_transform(data_txt['marital_status'])
data_txt['occupation']=LabelEncoder().fit_transform(data_txt['occupation'])
data_txt['relationship']=LabelEncoder().fit_transform(data_txt['relationship'])
data_txt['race']=LabelEncoder().fit_transform(data_txt['race'])
data_txt['sex']=LabelEncoder().fit_transform(data_txt['sex'])
data_txt['country']=LabelEncoder().fit_transform(data_txt['country'])
data_txt['class_income']=LabelEncoder().fit_transform(data_txt['class_income'])
```

```python
data_txt.head(6)
```

```python
X=np.array(data_txt.drop(['class_income'],1))
Y=np.array(data_txt['class_income'])
x_train,x_test,y_train,y_test=cross_validation.train_test_split(X,Y,test_size=0.3)
```

```python
print (type(x_train))
print (type(y_train))
print (x_train.shape)
print (y_train.shape)
```

```python
#make a classifier
#sklearn accepts input in terms of np_array
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
accuracy=clf.score(x_test,y_test)
print (accuracy)
#problem  is my data has strings do every kind 
#of list needs to be converted to float values  

#class income  0-<=50k  1->50k
```
